```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapz, simps #Importing trapz, and simps module from scipy for numerical integration
```

```python
pmf_dep = pd.read_csv('1-DEP.xvg', header = None, delim_whitespace = True)
pmf_dehp = pd.read_csv('2-DEHP.xvg', header = None, delim_whitespace=True)
pmf_didp = pd.read_csv('3-DIDP.xvg', header = None, delim_whitespace= True)
```

```python
df = pd.DataFrame({0: 0.000000, 1: [pmf_dep[1][0]], 2: [pmf_dep[2][0]]})
df = df.append(pmf_dep)
df = df.reset_index(drop = True)
pmf_dep = df.copy(deep = True)
pmf_dep = pmf_dep[pmf_dep[0] <= 4.000000]
pmf_dep = pmf_dep.append({0: 4.000000, 1: pmf_dep[1].iloc[-1] , 2: pmf_dep[2].iloc[-1]}, ignore_index=True)
pmf_dep[3] = np.exp(pmf_dep[1]/2.478)
```

```python
df = pd.DataFrame({0: 0.000000, 1: [pmf_dehp[1][0]], 2: [pmf_dehp[2][0]]})
df = df.append(pmf_dehp)
df = df.reset_index(drop = True)
pmf_dehp = df.copy(deep = True)
pmf_dehp = pmf_dehp[pmf_dehp[0] <= 4.000000]
pmf_dehp = pmf_dehp.append({0: 4.000000, 1: pmf_dehp[1].iloc[-1] , 2: pmf_dehp[2].iloc[-1]}, ignore_index=True)
pmf_dehp[3] = np.exp(pmf_dehp[1]/2.478)
```

```python
df = pd.DataFrame({0: 0.000000, 1: [pmf_didp[1][0]], 2: [pmf_didp[2][0]]})
df = df.append(pmf_didp)
df = df.reset_index(drop = True)
pmf_didp = df.copy(deep = True)
pmf_didp = pmf_didp[pmf_didp[0] <= 4.000000]
pmf_didp = pmf_didp.append({0: 4.000000, 1: pmf_didp[1].iloc[-1] , 2: pmf_didp[2].iloc[-1]}, ignore_index=True)
pmf_didp[3] = np.exp(pmf_didp[1]/2.478)
```

```python
plt.figure(figsize = (10,8))
plt.subplot(1,3,1)
plt.plot(pmf_dep[0], pmf_dep[1], color = 'r', label = 'DEP')
plt.plot(-pmf_dep[0], pmf_dep[1], color = 'r')
plt.ylim([-60,45])
plt.legend()
plt.subplot(1,3,2)
plt.plot(pmf_dehp[0], pmf_dehp[1], color = 'r', label = 'DEHP')
plt.plot(-pmf_dehp[0], pmf_dehp[1], color = 'r')
plt.ylim([-60,45])
plt.legend()
plt.subplot(1,3,3)
plt.plot(pmf_didp[0], pmf_didp[1], color = 'r', label = 'DIDP')
plt.plot(-pmf_didp[0], pmf_didp[1], color = 'r')
plt.ylim([-60,45])
plt.legend()
plt.tight_layout()
```
